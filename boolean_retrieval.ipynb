```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer
from nltk.tokenize import word_tokenize
import glob
import re
import os
import numpy as np
import sys
import zipfile as zf
stop_li=stopwords.words('english')
lemmatizer = WordNetLemmatizer()
```

```python
class Linked_LI:
    def __init__(self,head=None):
        self.head=head
```

```python
class Node:
    def __init__(self,doc_name,freq):
        self.doc_name=doc_name
        self.freq=freq
        self.next=None
```

```python
ultimate_words=[]
with zf.ZipFile("text_binary.zip", "r") as f:
    for name in f.namelist():
        data = f.read(name)
        data=data.decode("utf-8")
        data=re.sub(regex,'',data)
        words=word_tokenize(data.lower())
        words = [lemmatizer.lemmatize(word) for word in words if word not in stop_li]
        for word in words:
            if word not in ultimate_words:
                ultimate_words.append(word)
print(ultimate_words.__len__())
```

```python
dict_for_ll={}
for word in ultimate_words:
    dict_for_ll[word] = Linked_LI()
    dict_for_ll[word].head = Node(1,Node)
```
